**Import required Libraries**

```python
import requests
import gzip
from io import BytesIO
import xml.etree.ElementTree as ET
import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup
```

**Get Sitemap URLS**

example - https://www.flipkart.com/robots.txt

```python
def get_products_sitemap(website_url , contains_ = 'product'):
  robots_url = f'{website_url}/robots.txt'
  response = requests.get(robots_url)
  robots_content = response.text
  product_sitemap_directives = [line.split(': ')[1] for line in robots_content.split('\n') if line.startswith('Sitemap:') and contains_ in line]
  for sitemap_url in product_sitemap_directives:

    sitemap_response = requests.get(sitemap_url)
    sitemap_content = sitemap_response.content

    # Parse sitemap XML
    sitemap_root = ET.fromstring(sitemap_content)

    # Extract URLs from the sitemap XML
    urls = [url.text for url in sitemap_root.findall(".//{http://www.sitemaps.org/schemas/sitemap/0.9}loc")]

    # Process the extracted URLs as needed
    for url in urls:
        print(url)
    return urls

def getProductsURLs(sitemap_xml_gz_urls):
    anotherurls = []
    for product_url in sitemap_xml_gz_urls:
      print("fetching {}".format(product_url))
      response = requests.get(product_url)
      decompressed_data = gzip.GzipFile(fileobj=BytesIO(response.content)).read()
      # Parse the XML content
      root = ET.fromstring(decompressed_data)

      anotherurls = anotherurls + [url.text for url in root.findall(".//{http://www.sitemaps.org/schemas/sitemap/0.9}loc")]
    return anotherurls

def get_product_name(product_url, identifier, class_name):
  product_response = requests.get(product_url)
  product_soup = BeautifulSoup(product_response.content, 'html.parser')
  product_name = product_soup.find(identifier, class_= class_name).text
  print("Product Name:", product_name)
  return product_name
```

```python
website_url = 'https://www.flipkart.com'

urls = get_products_sitemap(website_url)
```

```python
c = getProductsURLs(urls)
print(c)
```

```python
def get_product_text(product_url):
  product_response = requests.get(product_url)
  product_soup = BeautifulSoup(product_response.content, 'html.parser')
  product_name_element = product_soup.find('span', class_= 'B_NuCI')
  product_name = product_name_element.text if product_name_element else "N/A"

  # Extract brand with null check
  brand_element = product_soup.find('span', class_='G6XhRU')
  brand = brand_element.text if brand_element else "N/A"

  # Extract category with null check
  category_elements = product_soup.find_all('a', class_='_2whKao')
  # Combine category text with "|" separator
  category_text = " | ".join([element.text for element in category_elements])
  return f"<name> {product_name} <brand> {brand} <category> {category_text}"
```

```python
all_products=[]
for product_url in c:
  text = get_product_text(product_url)
  all_products.append(text)
```
